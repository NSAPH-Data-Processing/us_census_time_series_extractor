```python
import duckdb
conn = duckdb.connect()
```

```python
conn.execute("""
    SELECT 
        year,
        SUM(pop_hispanic) AS pop_hispanic
    FROM
        'data/core/input/county__acs1__pop_hispanic.parquet'
    GROUP BY
        year
""").fetchdf()
```

```python
conn.execute("""
    SELECT 
        year,
        SUM(pop_hispanic) AS pop_hispanic
    FROM
        'data/core/output/county_yearly/acs1_*.parquet'
    GROUP BY
        year
""").fetchdf()
```

```python
conn.execute("""
    SELECT 
        year,
        AVG(median_household_income) AS median_household_income
    FROM
        'data/core/output/county_yearly/acs1_*.parquet'
    GROUP BY
        year
""").fetchdf()
```

```python
conn.execute("""
    SELECT 
        year,
        AVG(median_home_value) AS median_home_value
    FROM
        'data/core/output/county_yearly/acs1_*.parquet'
    GROUP BY
        year
""").fetchdf()
```

```python
conn.execute("""
    SELECT 
        year,
        AVG(median_household_income/median_home_value) AS income_2_home_value_ratio
    FROM
        'data/core/output/county_yearly/acs1_*.parquet'
    GROUP BY
        year
""").fetchdf()
```
